```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/RASIL-RISAM/-covid-research-/main/all_training.csv")
df
```

```python
df.columns
```

```python
df.shape
```

```python
df = df.dropna()
df.shape
```

```python
normal_ranges = {
    'CK': (5, 25),  # Normal range for CK
    'CREA': (0.74, 1.35),  # Normal range for CREA in males
    'GGT': (7, 47),  # Normal range for GGT in males
}
```

```python
for param, (lower, upper) in normal_ranges.items():
    df[f'{param}_Normal'] = ((df[param] >= lower) & (df[param] <= upper)).astype(int)

# Calculate ratios or differences between pairs of parameters
df['CK_CREARatio'] = df['CK'] / df['CREA']
df['CREA_GLURatio'] = df['CREA'] / df['GLU']
```

```python
df.head()
```

```python
x=df[['Age','CK_CREARatio','CREA_GLURatio','ALP','AST', 'ALT','LDH','RBC',
       'MCV','MO','BA','EOT']]
x
```

```python
y=df[['target']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.20, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
st_x= MinMaxScaler()
x_train= st_x.fit_transform(x_train)
x_test= st_x.transform(x_test)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train, y_train)
```

```python
y_pred= classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# Train the classifier
clf.fit(x_train, y_train)

# Get feature importances
feature_importances = clf.feature_importances_

# Create a DataFrame to display feature importances
feature_importance_df = pd.DataFrame({'Feature': x.columns, 'Importance': feature_importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print feature importances
print(feature_importance_df)
```

```python
from sklearn.model_selection import GridSearchCV

# Define a range of values for n_neighbors to be tested
param_grid = {'n_neighbors': np.arange(1, 21)}

# Initialize the KNN classifier
knn = KNeighborsClassifier()

# Perform grid search with cross-validation
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Evaluate the model with the best hyperparameters on the test set
best_knn = grid_search.best_estimator_
y_pred = best_knn.predict(x_test)
```

```python
# Print classification report and confusion matrix
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
